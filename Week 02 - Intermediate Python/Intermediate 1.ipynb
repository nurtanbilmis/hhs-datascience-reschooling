## Inleiding

In de volgende cel worden 5 lists aangemaakt:

1. turnover: omzet in GBP 
2. staff: aantal medewerkers
3. floor: winkeloppervlaktes
4. comp_sc: concurrentie score. Dit geeft een indruk van de concurrentiekracht van een winkel.
5. min30: aantal mensen dat op 30 minuten rijafstand woont van de betreffende winkel.

- Voer onderstaande cel uit.

```python
#Creating lists from columns in dataframe
import pandas as pd
import numpy as np

storedata = pd.read_csv('storedata_clean.csv', encoding = "ISO-8859-1",sep=";")

turnover = storedata['Profit'].values.tolist()

staff = storedata['Staff'].values.tolist()
floor = storedata['Floor Space'].values.tolist()
comp_sc = storedata['Competition score'].values.tolist()
min30 = storedata['30 min population'].values.tolist()
storedata.head()
```

```python
#import library
import matplotlib.pyplot as plt
```

```python
#scatter staff/turnover

# plt.scatter(staff, turnover)
plt.scatter(storedata['Staff'], storedata['Profit'])

plt.show()
```

```python
#correlation staff/turnover
np.corrcoef(staff, turnover)

storedata.corr()["Profit"] # storedata.corr() creates a new pandas dataframe
storedata.corr().loc["Profit"]
```

```python
#scatter floor/turnover
plt.scatter(floor, turnover)
plt.show()
```

```python
#correlation floor/turnover
np.corrcoef(floor, turnover)
```

```python
#scatter comp_sc/turnover
plt.scatter(comp_sc, turnover)
plt.show()
```

```python
#correlation comp_sc/turnover
np.corrcoef(comp_sc, turnover)
```

```python
#scatter min30/turnover
plt.scatter(min30, turnover)
plt.show()
```

```python
#correlation staff/turnover
np.corrcoef(staff, turnover)
```

```python
#histogram
plt.hist(turnover)

#number of bins and why: I think 5 makes sense to make it equal to tick amount
plt.hist(turnover, bins=5, color= "grey")

#lables and title
plt.title("Turnover Histogram")
plt.xlabel("Turnover")
plt.ylabel("Stores")

# definition of tick_val and tick_lab
tick_val = [1000000,2000000, 3000000,4000000, 5000000]
tick_lab = ["1M", "2M", "3M", "4M", "5M"]

# Adapt the ticks on the x-axis
plt.xticks(tick_val, tick_lab , rotation=30)

# After customizing, display the plot
plt.show()
```

```python
# #staafdiagram

# Wat staat er op de y-as: de gemiddelde omzet of de totale omzet?
# Ik denk gemiddelde omzet van de winkels met zelfde aantal medewerkers per column. Anders zou dit grafiek geen overzicht geven.

plt.bar(staff, turnover, color= "purple")
plt.title("Turnover(avg) by Staff Count")
plt.xlabel("Staff Count")
plt.ylabel("Turnover(avg)")
plt.yticks([1000000,2000000, 3000000,4000000, 5000000], ["1M", "2M", "3M", "4M", "5M"])
plt.show
```
