## 2 Control flow

### 2.1 if, else, elif

**stamford_staff**
- Print 'big store' als stamford_staff groter is dan 8. 
- Print 'medium store' als stamford_staff groter is dan 6 en kleiner of gelijk aan 8.
- Print 'small store' in alle andere gevallen.

**stamford_loc**
- Print stamford_loc als de locatie gelijk is aan 'Retail Park'. 
- Als dit niet het geval is print dan stamford_loc als de locatie gelijk is aan 'Shopping Centre'. 
- Print de eerste letter van de woorden in stamford_loc met hoofdletters als de eerdere checks niet waar zijn.

```python
stamford_staff = 8
stamford_loc = "shopping centre"


# Start coding here: stamford_staff
if stamford_staff > 8:
    print("big store")
elif 6 < stamford_staff <= 8:
    print("medium store")
else:
    print("small store")

    
# Start coding here: stamford_loc
if stamford_loc == "Retail Park":
    print(stamford_loc)
elif stamford_loc == "Shopping Centre":
    print(stamford_loc)
else:
    first, second = stamford_loc.split()
    print(first[0].upper(), second[0].upper())
    print(stamford_loc.title())
```
