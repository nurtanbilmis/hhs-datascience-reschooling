## Joining 2
In deze opdracht ga je aan de slag met transfergegevens van spelers uit de Engelse Premier League. Een transfer is de overgang van een speler van een club naar een andere club. Hierbij worden vaak enorme bedragen betaald.
- Voer onderstaande cel uit om 2 dataframes aan te maken:
    - pl18 bevat de gegevens van alle spelers die in het seizoen 2018-2019 in de Engels Premier League speelden
    - tr19 bevat de gegevens van alle transfers van en naar de Premier League gedurende het seizoen 2019-2020

```python
import pandas as pd
pl18_csv = pd.read_csv('england-premier-league-players-2018-to-2019-stats.csv')
tr19_csv = pd.read_csv('english_premier_league_2019_transfer.csv')
pl18 = pl18_csv.rename(columns={'Current Club': 'club_name'})
pl18 = pl18[['full_name', 'club_name']]
tr19 = tr19_csv[['player_name','club_name', 'club_involved_name', 'transfer_movement', 'fee']]
```

### 1 Inspecteren
Inspecteer beide dataframes:
- Laat zien welke kolommen er zijn en wat het datatype is van iedere kolom.
- Laat zien hoeveel rijen ieder dataframe heeft.
- Laat de eerste 5 rijen zien van ieder dataframe.
- Noteer hoeveel rijen je verwacht als je pl18 en tr19 samenvoegd met een 'right join'.
- Noteer hoeveel rijen je verwacht als je pl18 en tr19 samenvoegd met een 'left join'.

```python
display(tr19.columns, tr19.info(), tr19.shape)
print("________________________________________________\n")
display(pl18.columns, pl18.info(), pl18.shape)
print("________________________________________________\n")
print(pl18.head())
print(tr19.head())


# Noteer hoeveel rijen je verwacht als je pl18 en tr19 samenvoegd met een 'right join'.
# >>> 563

# Noteer hoeveel rijen je verwacht als je pl18 en tr19 samenvoegd met een 'left join'.
# >>> 572
```

### 2 inner join
- Voeg pl18 en tr19 samen in een nieuw dataframe op de naam van de speler met een inner join.
    - Aandachtspunt: de kolommen met de namen van de spelers zijn verschillend in beide dataframes.
    - Gebruik 'suffixes' om aan te geven uit welk dataframe een kolom afkomstig is.
- Hoeveel regels bevat het nieuwe dataframe?
- In hoeverre is dit een geschikte 'join' om alle transfers te kunnen analyseren?

```python
pl18_tr19 = pl18.merge(tr19, left_on = "full_name", right_on = "player_name", suffixes = ["_pl", "_tr"])
display(pl18_tr19.shape, pl18_tr19.head())

# In hoeverre is dit een geschikte 'join' om alle transfers te kunnen analyseren?
# >>> Niet.
```

### 3 Spelers zoeken
- Bepaal of je de gegevens van onderstaande spelers kunt terugvinden in het samengevoegde dataframe:
    - Rodri
    - Michy Batshuayi
- Waarom vind je de gegevens van de ene speler wel en van de ander speler niet?

```python
display(pl18_tr19[pl18_tr19["player_name"] == "Michy Batshuayi"])

display(pl18_tr19[pl18_tr19["player_name"] == "Rodri"])
# deze is niet overgenomen van de andere tabel
```

### 4 Left- en right joins
- Voeg pl18 en tr19 opnieuw samen maar zorg er nu voor dat alle regels uit tr19 in het nieuwe dataframe terecht komen.
- Toon de eerste 5 regels.
- Laat zien hoeveel regels het nieuwe dataframe heeft.
- Laat de gegevens zien van de speler Rodri.
- Van welke club naar welke club is hij getransfereerd? Wat was het transferbedrag?

```python
tr19.head()
pl18.head()
```

```python
pl18_tr19_2 = pd.merge(pl18, tr19, how = "right", left_on="full_name", right_on="player_name")
print(pl18_tr19_2.head())
print(pl18_tr19_2.shape)
display(pl18_tr19_2[pl18_tr19_2["player_name"] == "Rodri"])

#Van welke club naar welke club is hij getransfereerd? Wat was het transferbedrag?
# >>> Van Atletico Madrid naar Manchester City voor 63m GBP
```

## 5 Verwijderen van overbodige kolommen
- Verander de naam van kolom 'full_name' in pl18 in 'player_name'
    - Hint maak gebruik van: rename(columns={})
- Voeg pl18 en tr19 opnieuw samen zodat er in het nieuwe dataframe nog maar 1 kolom is met de naam van de spelers
- Gebruik 'suffixes' om aan te geven uit welk dataframe een kolom afkomstig is.
- Toon hoeveel rijen dit dataframe heeft
- Waarom is dit dataframe beter dan het vorige dataframe?

```python
pl18.rename(columns={"full_name":"player_name"}, inplace=True)
pl18_tr19_new = pd.merge(pl18, tr19, how="right", on="player_name", suffixes=("_18","_19"))
pl18_tr19_new.shape
pl18_tr19_new.head()
# een column voor spelernaam ziet er beter uit.
```

## 8 Analyse
- Maak een overzicht waarin je per club laat zien hoeveel 'inbound' en hoeveel 'outbound' transfers iedere club had.

```python
pl18_tr19_new.head()

transfer_mov = pl18_tr19_new.groupby(["club_name_19", "transfer_movement"])["player_name"].count()
display(transfer_mov)
```

### 9 Self-join
We gaan nu onderzoeken welke spelers naast een inbound transfer ook een outbound transfer hebben gemaakt. We maken hierbij gebruik van een self-join. 
- Voeg tr19 samen met tr19 op de naam van de speler in een nieuw dataframe
- Filter uit het nieuwe dataframe de regels met de volgende eigenschappen:
    - 'transfer_movement_x' is niet gelijk aan 'transfer_movement_y'
    - 'transfer_movement_x' is gelijk aan 'in'

```python
self_19 = tr19.merge(tr19, on = "player_name")
self_19.head()

self_19 = self_19[self_19["transfer_movement_x"] != self_19["transfer_movement_y"]]
self_19.head()

self_19 = self_19[self_19["transfer_movement_x"] == "in"]
display(self_19)
```
